```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact
import numpy as np
```

```python
dataset = "diamonds.csv"
df = pd.read_csv(dataset)
df.head()
```

```python
df.info()
```

```python
features = ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price', 'x', 'y', 'z']

for feature in features:
    print(f"Jumlah unique value for {feature}: {df[feature].nunique()}")
```

```python
# Mengelompokkan fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Mengelompokkan fitur kategorikal
categorical_features = ['cut', 'color', 'clarity']

# Mengelompokkan target
target = ['price']

# Menampilkan hasil pengelompokkan
print("Fitur Numerik:", numerical_features)
print("Fitur Kategorikal:", categorical_features)
print("Target:", target)
```

```python
# Membuat bar chart
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=feature, data=df, palette='viridis')
    plt.title(f'Bar Chart of {feature}', size=15)
    plt.xlabel(feature, size=12)
    plt.ylabel('Count', size=12)
    plt.show()
```

```python
# Membuat diagram lingkaran untuk setiap fitur kategorikal
for feature in categorical_features:
    plt.figure(figsize=(8, 6))  # Membuat gambar dengan ukuran tertentu
    df[feature].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'orange', 'lightblue'])
    plt.title(f'Diagram Lingkaran {feature}', size=15)  # Memberi judul diagram lingkaran
    plt.ylabel('')  # Menghapus label sumbu y
    plt.show()  # Menampilkan diagram lingkaran
```

```python
# Mengelompokkan fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Membuat distribusi untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))  # Membuat gambar dengan ukuran tertentu
    sns.histplot(df[feature], kde=True, color='darkcyan')  # Menggunakan sns.histplot untuk fitur numerik
    plt.title(f'Distribusi {feature}', size=15)  # Memberi judul distribusi
    plt.xlabel(feature, size=12)  # Label sumbu x
    plt.ylabel('Frekuensi', size=12)  # Label sumbu y
    plt.show()  # Menampilkan distribusi
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Menyiapkan subplots dengan 2 baris dan 3 kolom
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# Melakukan iterasi melalui fitur-fitur numerik
for i, feature in enumerate(numerical_features):
    # Membuat histogram menggunakan Seaborn
    sns.histplot(df[feature], kde=True, color='skyblue', ax=axes[i])
    axes[i].set_title(f'Histogram of {feature}')  # Menambahkan judul
    axes[i].set_xlabel(feature)  # Label sumbu x
    axes[i].set_ylabel('Frequency')  # Label sumbu y

# Mengatur tata letak subplot agar rapi
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for i, feature in enumerate(numerical_features):
    sns.boxplot(x=df[feature], ax=axes[i // 3, i % 3], orient='h', color='purple')
    axes[i // 3, i % 3].set_title(f'Boxplot of {feature}', size=12)
    axes[i // 3, i % 3].set_xlabel(feature, size=10)

#menyusun tata letak
plt.tight_layout()
plt.show()
```
