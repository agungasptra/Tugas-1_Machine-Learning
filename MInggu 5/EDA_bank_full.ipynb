```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "bank-full.csv"
df = pd.read_csv(dataset)
df.head()
```

```python
sns.countplot( x = df['age'] )
```

```python
sns.distplot(df['balance'])
```

```python
sns.boxplot(df['balance'])
```

```python
sns.scatterplot(x= df['age'], y = df['day']  , hue = df['balance'] , style = df['duration'] , size = df['campaign'])
plt.show ()
df['age'].min()
```

```python
sns.barplot (x = df['duration'] , y = df['balance'] , hue = df['duration'])
```

```python
sns.boxplot( x = df['duration'] , y = df['duration'] , hue = df['Day'] )
```

```python
import warnings
warnings.filterwarnings('ignore')
df['balance'].mean()

sns.distplot( df[df['duration'] ==  'Right' ]['balance'], hist = False) #pdf of that person  who is righty
sns.distplot( df[df['duration'] ==  'Left' ]['balance'] , hist = False) #pdf of that person  who is lefty
```

```python
df['balance'] = pd.to_numeric(df['balance'], errors='coerce')
print(df['balance'].dtype)

skill_X_preferredfoot  = pd.crosstab( df['duration'], df['duration'])
sns.heatmap(skill_X_preferredfoot)
plt.show()

new = df.groupby('duration')['balance'].mean()
new.plot(kind='bar')
plt.show()
```

```python
cluster = pd.crosstab(df['campaign'] ,df['duration'] )
sns.clustermap(cluster)
```
